```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import plotly
```

```python
df = pd.read_csv('../data/SAE_hyperparam_results_decoder_unitnorm.csv')
```

```python
#plot 'hidden_size' vs 'MSE' at epoch '301'
plt.scatter(df[(df['hidden_size']==4096) & (df['epoch']==601)]['L0'], df[(df['hidden_size']==4096) & (df['epoch']==601)]['MSE'], marker='o');
plt.scatter(df[(df['hidden_size']==2048) & (df['epoch']==601)]['L0'], df[(df['hidden_size']==2048) & (df['epoch']==601)]['MSE'], marker='x');
plt.scatter(df[(df['hidden_size']==1024) & (df['epoch']==601)]['L0'], df[(df['hidden_size']==1024) & (df['epoch']==601)]['MSE'], marker='*');
#axis
plt.xlabel('L0')
plt.ylabel('MSE')
#legend
plt.legend(['4096', '2048', '1024'])
#xlim
plt.xlim(11, 17)
#ylim
plt.ylim(0, .01)
```
